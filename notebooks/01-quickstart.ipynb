## データディレクトリの所在

`aiseed-weather` は ECMWF / JMA / Open-Meteo のキャッシュをひとつのディレクトリにまとめている。アプリの設定 (`~/.config/aiseed-weather/config.toml` の `data_dir`) があればそれを、無ければ `platformdirs.user_cache_dir("aiseed-weather")` を使う。

````python
from aiseed_weather.models.user_settings import load_or_init, resolved_data_dir

settings = load_or_init().settings
data_dir = resolved_data_dir(settings)
print(f"data_dir = {data_dir}")
print(f"exists   = {data_dir.exists()}")
````

````python
# キャッシュ内のトップレベルを覗く
if data_dir.exists():
    for child in sorted(data_dir.iterdir()):
        marker = "DIR " if child.is_dir() else "FILE"
        print(f"  {marker}  {child.name}")
else:
    print("(まだ空。一度 `flet run` でアプリを起動して、地点を 1 つ追加 → 更新ボタンで予報を取ると ECMWF/Open-Meteo のキャッシュが生成されます)")
````

## 保存済み地点を確認


````python
from aiseed_weather.models.point_location import load_locations

locations = load_locations(data_dir)
if locations:
    for loc in locations:
        print(f"  {loc.name:20s} ({loc.latitude:.3f}, {loc.longitude:.3f})  tz={loc.timezone}")
else:
    print("地点はまだ保存されていません。アプリで追加するか、このノート内で `Location.new(...)` を直接構築してください。")
````
